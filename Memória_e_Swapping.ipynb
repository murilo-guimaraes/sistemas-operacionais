##**Introdução Teórica**

O gerenciamento de memória é uma das funções mais complexas do Sistema Operacional, responsável por coordenar o uso da memória principal (RAM) entre os diversos processos. Segundo Barbosa (2018), quando a memória física se torna insuficiente para manter todos os processos ativos, o SO utiliza a técnica de **Swapping**. O Swapping consiste em mover temporariamente um processo da memória principal para a memória secundária (disco) e trazê-lo de volta quando necessário. Essa gerência é vital para permitir a multiprogramação e garantir que cada programa possua seu próprio espaço de endereçamento isolado e seguro.

### **1. Diagnóstico de Memória Real e Swap**

Nesta etapa, utilizaremos comandos de sistema para verificar a situação da memória no ambiente Linux do Colab. Como explica Barbosa, o SO monitora constantemente quanto de RAM está disponível e quanto de espaço de Swap (disco) está sendo utilizado para sustentar a carga de trabalho atual.

```python
import matplotlib.pyplot as plt
import subprocess
import re

# Executa o comando free e captura a saída
cmd = subprocess.check_output(['free', '-m']).decode('utf-8')
linhas = cmd.split('\n')
dados_mem = re.findall(r'\d+', linhas[1])

# Extraindo valores: total, usada, livre, buff/cache
total, usada, livre, compart, buff_cache, disponivel = map(int, dados_mem)

# Criando o gráfico de pizza (Donut Chart)
labels = ['Usada', 'Livre', 'Buff/Cache']
valores = [usada, livre, buff_cache]
cores = ['#ff9999','#66b3ff','#99ff99']

plt.figure(figsize=(7, 7))
plt.pie(valores, labels=labels, autopct='%1.1f%%', startangle=140, colors=cores, pctdistance=0.85)

# Desenha o círculo central para parecer um donut
centro = plt.Circle((0,0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centro)

plt.title(f'Distribuição da Memória RAM Real (Total: {total}MB)')
plt.show()

print(f"Detalhes técnicos: Total {total}MB | Usada {usada}MB | Livre {livre}MB")
```

#### **O que observar:**
<a>O gráfico de pizza permite visualizar a "fatia" de memória que o Sistema Operacional reserva para cache e buffers, além da memória livre. Conforme Barbosa explica na Unidade 3, o SO tenta manter o máximo de dados possível em cache para acelerar o sistema, mas precisa gerenciar a parte "Livre" para evitar o acionamento do Swapping.
>**Nota Técnica**: A visualização em tempo real ajuda a identificar se o ambiente está próximo do limite crítico onde o "Swap-Out" (movimentação para o disco) se tornaria obrigatório para manter a estabilidade.</a>

### **2. Simulação de Alocação e Alerta de Limite**

Para entender como o gerenciador de memória atua, criaremos um script que aloca memória gradualmente. Isso simula o comportamento de um processo que cresce dinamicamente e exige que o SO decida entre manter os dados na RAM ou realizar o swapping, conforme descrito por Barbosa.

```python
import sys

# Criando uma lista para ocupar memória de forma controlada
print("[*] Iniciando simulação de alocação de memória...")
lista_memoria = []

try:
    for i in range(1, 6):
        # Aloca aproximadamente 100MB por passo
        bloco = ' ' * (100 * 1024 * 1024)
        lista_memoria.append(bloco)
        print(f"Passo {i}: Alocados ~{i*100}MB")
except MemoryError:
    print("[ALERTA] Limite de memória física atingido! SO acionando Swapping.")

print("\n[OK] Simulação de alocação finalizada.")
```

#### **O que observar:**
<a>Neste teste, observamos a tentativa do Python de reservar grandes blocos de memória. Se o sistema não interromper o processo com um erro, ele moverá outros processos inativos para o disco.
> **Nota Técnica:** No Colab, o ambiente é monitorado; se excedermos muito o limite, o kernel é reiniciado. Isso ilustra o conceito de proteção de memória e gerência de recursos críticos.</a>

### **3. Lógica de Swapping: Troca de Processos**

O Swapping remove processos inteiros da memória para o disco para liberar espaço. Simularemos essa lógica de "In" e "Out" (entrada e saída), demonstrando como o SO mantém uma tabela de controle para saber onde cada parte do processo está localizada (se na RAM ou no arquivo de swap).

```python
import matplotlib.pyplot as plt

# Estado da Simulação
ram = ["Processo_C", "Processo_B"]
swap = ["Processo_A"]

plt.figure(figsize=(10, 2))
plt.barh(["RAM"], [len(ram)], color='lightgreen', label='Na Memória')
plt.barh(["Disco (Swap)"], [len(swap)], color='salmon', label='Em Swap')

plt.title('Simulação de Localização de Processos (Swapping)')
plt.xlabel('Quantidade de Processos')
plt.legend()
plt.show()
```

#### **O que observar:**
<a>A simulação demonstra a troca de contexto de memória. Quando o "Processo_C" precisa entrar, o SO escolhe um processo (geralmente o menos usado) para sair.
>**Nota Técnica:** Na prática, o Swapping de processos inteiros é caro para o desempenho (I/O de disco); por isso, sistemas modernos preferem o "Paging" (paginação), mas o conceito de troca de base permanece o mesmo descrito por Barbosa na Unidade 3.</a>
